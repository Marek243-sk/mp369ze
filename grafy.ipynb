```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataBc.csv") # Načítanie datasetu

# Premenovanie atribútov
df = df.rename(columns={"credit_score":"Kreditné skóre", "country":"Krajina", "gender":"Pohlavie", "age":"Vek", "tenure":"Dĺžka existencie účtu", 
                        "balance":"Zostatok na účte", "products_number":"Počet produktov", "credit_card":"Kreditná karta", 
                        "active_member":"Aktivita zákazníka", "estimated_salary":"Odhadovaný plat", "churn":"Odchodovosť"})

# Premenovanie hodnôt v atribútoch 'Krajina' a 'Pohlavie'
df["Krajina"] = df["Krajina"].replace(["France", "Germany", "Spain"],["Francúzsko", "Nemecko", "Španielsko"])
df["Pohlavie"] = df["Pohlavie"].replace(["Female", "Male"],["Žena", "Muž"])

pd.concat([df.head(5), df.tail(5)]) # Zobrazenie 5 riadkov zo začiatku a konca datasetu
```

Nominálne atribúty

```python
# nastavenie mriežky pre grafy
fig, axs = plt.subplots(2, 1, constrained_layout=True, figsize=(5, 7))
sns.set_context("notebook", font_scale=1.1)

# Prvý graf pre atribút 'Krajina'
sns.countplot(data=df, x="Krajina", hue="Krajina", palette="dark:blue", legend=False, ax=axs[0])
axs[0].set_title("Zákazníci podľa krajina pôvodu")
axs[0].set_ylabel("Počet")

# Druhý graf pre atribút 'Pohlavie'
sns.countplot(data=df, x="Pohlavie", hue="Pohlavie", palette="dark:blue", legend=False, ax=axs[1])
axs[1].set_title("Zákazníci podľa pohlavia")
axs[1].set_ylabel("Počet")
```

```python
# Nastavenie mriežky pre grafy
fig, axs = plt.subplots(2, 2, constrained_layout=True, figsize=(8, 8))
sns.set_context("notebook", font_scale=1.2)

# Prvý graf pre atribút 'Kreditná karta'
sns.countplot(data=df, x="Kreditná karta", hue="Kreditná karta", legend=False, palette="dark:blue", ax=axs[0,0])
axs[0,0].set_title("Zákazníci podľa vlastníctva kreditnej karty")
axs[0,0].set_ylabel("Počet")

# Druhý graf pre atribút 'Aktivita zákazníka'
sns.countplot(data=df, x="Aktivita zákazníka", hue="Aktivita zákazníka", legend=False, palette="dark:blue", ax=axs[0,1])
axs[0,1].set_title("Zákazníci podľa aktivity")
axs[0,1].set_ylabel("Počet")

# Tretí graf pre atribút 'Odchodovosť'
sns.countplot(data=df, x="Odchodovosť", hue="Odchodovosť", legend=False, palette="dark:blue", ax=axs[1,0])
axs[1,0].set_title("Zákazníci podľa toho, či opustili banku")
axs[1,0].set_ylabel("Počet")

# Vypnutie vykresľovania pre posledné miesto mriežky
axs[1,1].axis("off")
plt.show()
```

# Druhá fáza CRISP-DM, grafy - > box, numerické

```python
# Nastavenie mriežky pre grafy
fig, axs = plt.subplots(2, 1,constrained_layout=True, figsize=(10, 8))

# Prvý graf pre atribút 'Kreditné skóre'
sns.boxplot(data=df, x="Kreditné skóre", ax=axs[0])
axs[0].set_title("Zákazníci podľa kreditného skóre")
axs[0].set_xlabel("Vyška kreditného skóre")

# Druhý graf pre atribút 'Vek'
sns.boxplot(data=df, x="Vek", ax=axs[1])
axs[1].set_title("Zákazníci podľa veku")
axs[1].set_xlabel("Vek")
```

```python
# Nastavenie mriežky pre grafy
fig, axs = plt.subplots(2, 1, constrained_layout=True, figsize=(10, 8))

# Prvý graf pre atribút 'Dĺžka existencie účtu'
sns.boxplot(data=df, x="Dĺžka existencie účtu", ax=axs[0])
axs[0].set_title("Dĺžka existencie účtu")
axs[0].set_xlabel("Roky")

# Druhý graf pre atribút 'Zostatok na účte'
sns.boxplot(data=df, x="Zostatok na účte", ax=axs[1])
axs[1].set_title("Zostatok na účte zákazníkov")
axs[1].set_xlabel("Suma v €")
```

```python
# Nastavenie mriežky pre grafy
fig, axs = plt.subplots(2, 1, constrained_layout=True, figsize=(10, 8))

# Prvý graf pre atribút 'Počet produktov'
sns.boxplot(data=df, x="Počet produktov", ax=axs[0])
axs[0].set_title("Počet produktov využívaných zákazníkmi")
axs[0].set_xlabel("Počet produktov")

# Druhý graf pre atribút 'Odhadovaný plat'
sns.boxplot(data=df, x="Odhadovaný plat", ax=axs[1])
axs[1].set_title("Odhadovaný plat zákazníkov")
axs[1].set_xlabel("Plat v €")
```

```python
# Selekcia numerických atribútov datasetu do premennej
df_num = df[["Kreditné skóre", "Vek", "Dĺžka existencie účtu", "Zostatok na účte", "Počet produktov", "Odhadovaný plat"]]

# Vytvorenie korelačnej matice
correlation_matrix = df_num.corr()

# Vykreslenie korelačnej matice
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, annot_kws={"size": 12}, linecolor="black", cmap="Blues")
```
